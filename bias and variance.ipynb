```python
import numpy as np
import matplotlib.pyplot as plt

# Example datas
x = np.linspace(0, 1, 15)
y = np.array([1.5, 1.6, 1.7, 2.30, 2.15, 2.25, 2.50, 1.77, 1.60, 1.88, 1.75, 1.07,1.0, 0.90, 0.20])
plt.scatter(x, y)
```

```python
p_1 = np.polyfit(x, y, 1)
plt.scatter(x, y)
plt.plot(x,np.polyval(p_1, x), c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Underfitting & High Bias')
plt.legend()
```

```python
p_2 = np.polyfit(x, y, 15)
plt.scatter(x, y)
plt.plot(x,np.polyval(p_2, x), c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Overfitting & High Variance')
plt.legend()
plt.savefig('overfitting.png')
```

```python
p_2 = np.polyfit(x, y, 3)
plt.scatter(x, y)
plt.plot(x,np.polyval(p_2, x), c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic model that seems to be good')
plt.legend()
```
